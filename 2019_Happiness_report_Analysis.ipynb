```python
#calculate Outliers
def calculate_outliers(column):
    q3=np.quantile(df[column],q=0.75)
    q1=np.quantile(df[column],q=0.25)
    IQR=q3-q1
    lower_inbound=q1-(IQR*1.5)
    Upper_inbound=q3+(IQR*1.5)

    return df[(df[column]<lower_inbound) | (df[column]>Upper_inbound)][column]
```

```python
print(calculate_outliers('Social support'))
```

```python
print(calculate_outliers('Perceptions of corruption'))  #total 14 outliers in the Perceptions of Corruptions
```

```python
print(calculate_outliers('Freedom to make life choices'))
```

```python
print(calculate_outliers('Generosity'))
```

```python
#is any Column  is Normally Distributed ?

def Normal(column):
    dict={}
    mean=df[column].mean()
    std=df[column].std()
    std_1_pos=mean+std
    std_1_neg=mean-std
    count=len(df[(df[column]>=std_1_neg) & (df[column]<std_1_pos)][column])
    dict['68%']=(count/len(df[column]))*100
    std_2_pos=mean+2*std
    std_2_neg=mean-2*std
    dict['95%']=(len(df[(df[column]>=std_2_neg) & (df[column]<std_2_pos)][column])/len(df[column]))*100
    std_3_pos=mean+3*std
    std_3_neg=mean-3*std
    dict['99.7%']=(len(df[(df[column]>=std_3_neg) & (df[column]<std_3_pos)][column])/len(df[column]))*100
    dict['variance']=np.var(df[column])
    return dict
```

```python
print(Normal('Score'))
```

```python
print(Normal('Healthy life expectancy'))
```

```python
print(Normal('GDP per capita'))
```

```python
print(Normal('Generosity'))
```

```python
print(Normal('Perceptions of corruption'))
```
